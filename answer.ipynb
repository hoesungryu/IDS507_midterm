```python
import os
import numpy as np
import pandas as pd 

import missingno as msno
```

```python
LOW_HZ= [f'{10*i}HZ' for i in range(1,25)]
HIGH_HZ = [f'{4000+10*i}HZ' for i in range(50)]
max_bin = [f'max{i}' for i in range(20)]
```

## Pre-processing


```python
def convert2str(x):
    int_dict = {'low':0,'mid':1,'high':2}
    return int_dict[x]
```

```python
# low, mid, high ---> 2
df[max_bin] = df[max_bin].applymap(convert2str)
```

```python
# 결측치 채우기 ----> 
df[LOW_HZ] = df[LOW_HZ].interpolate(method='linear', limit_direction='both',axis=1)
```

```python
df
```

```python
# conver string --> float
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler

scaler = MinMaxScaler() # scaler 정의 |
scaler.fit(df[HIGH_HZ])

df[HIGH_HZ] = scaler.transform(df[HIGH_HZ])
```

```python
# split into train  & test
from sklearn.model_selection import train_test_split

X = df.drop(['label'],axis=1)
y = df.label


X_train, X_valid, y_train, y_valid  = train_test_split(X,y, test_size=0.2 ,stratify=y, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X_train,y_train)


y_pred = lr.predict(X_valid)
pred_proba = lr.predict_proba(X_valid)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score


# ACC 
accuracy = accuracy_score(y_valid, y_pred)


# calculate AUC of model
roc_auc = roc_auc_score(y_valid, pred_proba[:,1]) # calculate AUC of model

print("Accuracy : ",round(accuracy,4))
print('AUC score:', round(roc_auc,4) )
```

```python
import pandas as pd 
import seaborn as sns 
from sklearn.metrics import confusion_matrix


confusion = confusion_matrix(y_valid, y_pred)

matrix = pd.DataFrame(confusion,
                     columns = ['Positive','Negative'],
                     index= ['True','False']
                     )
sns.heatmap(matrix, annot=True, cmap='Blues', fmt='d')
```

```python
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
# plt.style.use('seaborn')


# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_valid, pred_proba[:,1], pos_label=1)


# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='Logistic Regression')
# plt.plot(fpr2, tpr2, linestyle='--',color='green', label='KNN')
# plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
plt.plot([0,1],[0,1],linestyle='--', color='blue')
# title
plt.title('ROC curve for Classification')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
# plt.savefig('ROC',dpi=300)
plt.show();

```

```python
thresholds = [0.3 , 0.33 ,0.36,0.39, 0.42 , 0.45 ,0.48, 0.50]

pred_proba_c1 = pred_proba[:,1].reshape(-1,1)
```

```python
from sklearn.preprocessing import Binarizer


for custom_threshold in thresholds:
    binarizer = Binarizer(threshold=custom_threshold).fit(pred_proba_c1) 
    custom_predict = binarizer.transform(pred_proba_c1)
    print('Threshold:',custom_threshold)
    accuracy = accuracy_score(y_valid , custom_predict)
    print("Accuracy: ",round(accuracy,3))
    print(" ")
```
